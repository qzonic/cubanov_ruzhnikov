Построить нормированную амплитудную диаграмму направленности полного вектора напряженности электрического поля элементарного
электрического излучателя в плоскости XOξ при $r=1000 м$, φ=45°. Излучатель ориентирован вдоль оси Х (рис. 2.27). Амплитуда тока, возбуждающего излучатель, равна 1 А, частота 300 МГц, длина излучателя 10 см.

![](../img/Chapter_2_example_1_picture_1.jpg)

**Решение задачи**

```python
# Импортируем необходиую библиотеку
import numpy
import matplotlib.pyplot as plt
```

```python
# Коэффициент для перевода из радиан в градусы
RAD2DEG = numpy.pi/180
# Зададим азимутный угол
PHI = RAD2DEG * numpy.linspace(45, 45, 360)
# Задаем угл тета в градусах с параметрами:
# начальное значение - 0.01
# конечное значение - 360
# количество точек - 360
TETA = RAD2DEG * numpy.linspace(0.01, 360, 360)
```

```python
# Функция get_unnormalized_hn рассчитывает ненормированную характеристику направленности в зависимости от угла teta.
def get_unnormalized_hn():
    return numpy.array(numpy.abs(numpy.sqrt(1 - numpy.cos(PHI)**2*numpy.sin(TETA)**2)))


# Функция get_max_hn рассчитывает максимальное значение ненормированной характеристики направленности
def get_max_hn(hn):
    return numpy.max(hn)


# Функция get_normalized_hn рассчитывает ненормированную характеристику направленности,
# определяет ее максимальное значение,
# а потом рассчитывает нормированную характеристику направленности
def get_normalized_hn():
    f = get_unnormalized_hn()
    M = get_max_hn(f)
    return f/M
```

Определим нормированную характеристику направленности

```python
F = get_normalized_hn()
```

```python
plt.polar(TETA, F)
plt.show()
```

```python
plt.plot(TETA / RAD2DEG, F)
plt.show()
```
