Прямоугольная поверхность (рис. 5.19) находится в центре системы координат и имеет размер $a = 4λ, b = 8λ$. Распределение фазы возбуждения по оси X равномерное, а по оси Y имеет вид $ψ_1(y) = −πy⁄b$ (рис. 5.18, а). Распределение амплитуды возбуждения по осям X и Y — равномерное. Рассчитать нормированную амплитудную характеристику направленности возбужденной поверхности в плоскости ZOY в пределах −40° ... 40°и построить её диаграммы в прямоугольной системе координат с линейным и логарифмическим масштабом по оси ординат.

```python
# Импортируем необходимые библиотеки
import numpy
import matplotlib.pyplot as plt
```

```python
a = 4  # размер вдоль оси x в долях длины волны
b = 8  # размер вдоль оси y в долях длины волны
psi = numpy.pi/2  # максимальное значение фазы
# Задаем угл тета в градусах с параметрами:
# начальное значение - 0.01
# конечное значение - 360
# количество точек - 360
TETA = (numpy.pi/360)*numpy.linspace(0.01, 360, 360)
```

```python
# Функция get_unnormalized_hn рассчитывает ненормированную характеристику направленности.
def get_unnormalized_hn(teta):
    return numpy.array(
        (1 + numpy.cos(teta)) *
        (numpy.sin((b*(2 * numpy.pi * numpy.sin(teta))/2)-psi)/
         ((b*(2 * numpy.pi * numpy.sin(teta))/2)-psi))
    )


# Функция get_max_hn рассчитывает максимальное значение ненормированной характеристики направленности
def get_max_hn(hn):
    return numpy.max(hn)


# Функция get_normalized_hn рассчитывает нормированную характеристику направленности
def get_normalized_hn(teta, maxf):
    f = get_unnormalized_hn(teta)
    return (1/maxf) * numpy.abs(f)
```

```python
f = get_unnormalized_hn(TETA)
M = get_max_hn(f)
print('M =', M)
```

```python
# Задаем угл тета в градусах с параметрами:
# начальное значение - -40
# конечное значение - 40
# количество точек - 360
TETA = (numpy.pi/180)*numpy.linspace(-40, 40, 360)
F = get_normalized_hn(TETA, M)  # Получаем нормированную диаграмму направленности
```

```python
plt.plot(TETA / (numpy.pi/180), F)
plt.grid()
plt.show()
```

```python
F = 20 * numpy.log10(F)
plt.plot(TETA / (numpy.pi/180), F)
plt.grid()
plt.show()
```
