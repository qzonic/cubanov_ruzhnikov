Задана система двух линейных симмет- ричных электрических вибраторов с длиной плеча $l = 0,4λ$, показанных на рис. 3.25. Расстояние $d = 1,0λ$, отношение комплексных значений токов $I_2/I_1 = e^{\frac{jπ}{2}}$. Рассчитать нормированную амплитудную характери- стику направленности и построить нормированную амплитудную диаграмму направленности в плоскости YOZ в полярной системе координат.

![](../img/Chapter_3_example_4_picture_1.png)

$f(v)=\frac{cos(klsin(v)) - cos(kl)}{cos(v)}\sqrt{1 + q^2 + 2qcos(ψ-kdcos(v))}$ (1)

```python
# Импортируем необходимые библиотеки
import numpy
import matplotlib.pyplot as plt
```

```python
kl = 0.8*numpy.pi
kd = 2*numpy.pi
q = 1
# Задаем угл V в градусах с параметрами:
# начальное значение - 0.01
# конечное значение - 360
# количество точек - 360
V = (numpy.pi/180)*numpy.linspace(0.01, 360, 360)
psi = numpy.pi/2
```

```python
# Функция get_unnormalized_hn рассчитывает ненормированную амплитудную характеристику направленности
def get_unnormalized_hn(v):
    return numpy.array(((numpy.cos(kl*numpy.sin(v))-numpy.cos(kl))/numpy.cos(v))*numpy.sqrt(1+q**2+2*q*numpy.cos(psi - kd*numpy.cos(v))))

# Функция get_max_hn рассчитывает максимвльное значение ненормированной амплитудной характеристики направленности
def get_max_hn(hn):
    return numpy.max(hn)

# Функция get_normalized_hn рассчитывает нормированную амплитудную характеристику направленности
def get_normalized_hn(hn, maxhn):
    return numpy.abs(hn/maxhn)
```

```python
f = get_unnormalized_hn(V)  # Получаем ненормированную амплитудную характеристику направленности
M = get_max_hn(f)  # Получаем максимальное значение ненормированной амплитудной характеристики направленности
F = get_normalized_hn(f, M)
```

```python
plt.polar(V, F)
plt.show()
```
