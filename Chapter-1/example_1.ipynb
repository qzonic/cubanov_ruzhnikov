**Задача 1.** В системе координат, приведенной на рис. 1.24, характеристика направленности некоторой антенны описывается функцией

![](../img/Chapter_1_example_1_picture_1.jpg)

$f(θ,φ)=\frac{sin(3πsin(φ)sin(θ))}{1,5πsin(φ)sin(θ)}$  (1)

Для плоскости φ = *const* = 90° построить нормированную диаграмму направленности этой антенны в полярной системе координат и прямоугольной системе координат с логарифмическим масштабом.
**Решение задачи.**
Заданная характеристика направленности при φ = 90° зависит только от угла θ, следовательно, уравнение (1) упрощается:
*$f(θ)=\frac{sin(3πsinθ)}{1,5πsinθ}$*  (2)

```python
# Импортируем необходимые библиотеки
import numpy
import matplotlib.pyplot as plt
```

```python
# Задаем угл тета в радианах с параметрами:
# начальное значение - 0.01
# конечное значение - 360
# количество точек - 360
TETA = numpy.linspace(0.01, 360, 360)
# Коэффициент для перевода из радиан в градусы
RAD2DEG = numpy.pi/180
```

```python
# Функция get_unnormalized_hn рассчитывает ненормированную характеристику
# направленности в зависимости от угла teta по формуле 2.
def get_unnormalized_hn(teta):
    return numpy.array((numpy.sin(3 * numpy.pi * numpy.sin(teta))) / (1.5 * numpy.pi * numpy.sin(teta)))


# Функция get_normalized_hn рассчитывает нормированную характеристику направленности
def get_normalized_hn():
    f = get_unnormalized_hn(RAD2DEG * TETA)
    maxf = numpy.max(f)
    return (1/maxf) * numpy.abs(f)
```

```python
# Рассчитываем нормированную характеристику направленности
# и переводим ее в логарифмический масштаб
F = 20 * numpy.log10(get_normalized_hn())
plt.plot(TETA, F)  # Построение графика в прямоугольной системе координат
plt.grid()
plt.show()  # Отображение графика
```

```python
plt.polar(RAD2DEG * TETA, get_normalized_hn())  # Построение графика в полярной системе координат
plt.show()  # Отображение графика
```
