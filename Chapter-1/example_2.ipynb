**Задача 2.**
В системе координат, приведенной на рис. 1.24, нормированная характеристика направленности некоторой антенны описывается функцией
$F(θ)=\frac{cos(1,4πcos(θ))-cos(1,4π)}{(1-cos(1,4π))sin(θ)}$  (1)
Определить число боковых лепестков и их уровни в децибелах. Определить ширину главного лепестка диаграммы направленности по уровню нулевого излучения $2θ_{0}$ и по уровню половинной мощности $2θ_{0,5}$.

![](../img/Chapter_1_example_1_picture_1.jpg)

```python
# Импортируем необходимые библиотеки
import numpy
import matplotlib.pyplot as plt

numpy.seterr(divide = 'ignore')  # Отключаем исключение, которое возникает из-за деления на 0
```

```python
# Задаем угл тета в радианах с параметрами:
# начальное значение - 0.01
# конечное значение - 360
# количество точек - 360
TETA = numpy.linspace(0.01, 360, 360)
# Коэффициент для перевода из радиан в градусы
RAD2DEG = numpy.pi/180
```

```python
# Функция get_normalized_hn рассчитывает нормированную характеристику направленности
def get_normalized_hn(teta):
    return numpy.array(numpy.abs((numpy.cos(1.4*numpy.pi*numpy.cos(teta))-numpy.cos(1.4*numpy.pi))
                                 / ((1-numpy.cos(1.4*numpy.pi))*numpy.sin(teta))))
```

```python
plt.polar(RAD2DEG * TETA, get_normalized_hn(RAD2DEG * TETA))  # Построение графика в полярной системе координат
plt.title('Рис. 1')  # Задаем title для графика
plt.show()  # Отображение графика
```

По диаграмме определяем, что число боковых лепестков равно четырем.

```python
# Рассчитываем нормированную характеристику направленности
# и переводим ее в ллогарифмический масштаб
F = 20 * numpy.log10(get_normalized_hn(RAD2DEG * TETA))
plt.plot(TETA, F)  # Построение графика в прямоугольной системе координат
plt.grid(axis = 'y')  # Добавляем основныи линии (0, -10, -20, ...)
plt.minorticks_on()  # Включаем второстепенное деления осей
plt.grid(which='minor', linestyle = ':', axis = 'y')  # Определяем внешний вид линий вспомогательной сетки
plt.plot(TETA, numpy.linspace(-2,-2,360))  # Строим горизонтальную линию на уровне -2 дБ.
plt.title('Рис. 2')  # Задаем название для графика
plt.show()  # Отображение графика
```

По рисунку 2 удобно определить уровень боковых лепестков в децибелах: ξ = −2,0 дБ.

```python
plt.plot(TETA, get_normalized_hn(RAD2DEG * TETA))  # Построение графика в прямоугольной системе координат
plt.grid(axis = 'y')
plt.minorticks_on()  # Включаем второстепенное деления осей
plt.grid(which='minor', linestyle = ':', axis = 'y')  # Определяем внешний вид линий вспомогательной сетки

plt.plot(numpy.linspace(65,65,360), get_normalized_hn(RAD2DEG * TETA), '--', color='green')  # Строим вертикальную линию по OX на уровне 65°
plt.plot(numpy.linspace(115,115,360), get_normalized_hn(RAD2DEG * TETA), '--', color='green')  # Строим вертикальную линию по OX на уровне 115°

plt.plot(numpy.linspace(78,78,360), get_normalized_hn(RAD2DEG * TETA), '--', color='red')  # Строим вертикальную линию по OX на уровне 78°
plt.plot(numpy.linspace(102,102,360), get_normalized_hn(RAD2DEG * TETA), '--', color='red')  # Строим вертикальную линию по OX на уровне 102°

plt.title('Рис. 3')  # Задаем название для графика
plt.show()  # Отображение графика
```

На рис. 3 вновь приведена диаграмма F(θ), но построенная уже в прямоугольной системе координат с линейным масштабом по оси ординат. По этой диаграмме очень просто определить ширину диаграммы направленности по уровню нулевого излучения  $2θ_{0}=50°$ и по уровню половинной мощности $2θ_{0,5}=24°$
